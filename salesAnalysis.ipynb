```python
#to get each location
location_list= sales.groupby('Location')
location = [x for x,y in location_list]
location
```

```python
#to get sum of each transaction for each location
total=sales.groupby('Location').sum()['Total']
total
```

```python
import matplotlib.pyplot as plt
plt.bar(location,total)
plt.show()
```

```python
#representing data of each location on pie chart


plt.pie(total,labels=location,autopct='%1.1f%%')
plt.show()
```

```python
# mapping data based on gender and location
location_sales = sales.groupby(['Gender','Location']).count()['Invoice ID']
location_sales.unstack(level=0)
```

```python
unstacked_shopper=location_sales.unstack(level=0)
```

```python
unstacked_shopper.plot(kind='bar')
plt.show()
```

```python
spend=sales.groupby('Gender').sum()['Total']    #checking which gender spends more
spend
```

```python
spend.plot(kind='bar')            #plotting for gender spending more
plt.show()
```

```python
members = sales.groupby(['Member','Location']).count()['Invoice ID']
members.unstack(level=0)
```

```python
#creating new column with Day's

sales['Day']=pd.to_datetime(sales['Date']).dt.day
sales.head(10)
```

```python
#calculating total based on each day of each month
sales_day=sales.groupby('Day').sum()['Total']
sales_day
```

```python
sales_day.plot()
plt.show()
```

```python
sales['Hour']=pd.to_datetime(sales['Time']).dt.hour
sales.head(10)
```

```python
#calculating total based on each day of each month
sales_hour=sales.groupby('Hour').sum()['Total']
sales_hour
```

```python
sales_hour.plot()
plt.show()
```

```python
#finding which type of payment is used for each hour
payType_hour=sales.groupby(['Hour','Payment']).count()['Total']
payType_hour.unstack()
```

```python
uPay_Type=payType_hour.unstack()
uPay_Type.plot()
plt.show()
```
